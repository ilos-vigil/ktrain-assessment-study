```python
WIKIPEDIA_PATH = ',/dataset/txt'
N_DOCS_CONSIDERED = 8
QUESTION_TYPE = 'Soal' # "Soal" or "Soal_5W1H"
```

```python
import logging
import pandas as pd
from haystack.document_store.memory import InMemoryDocumentStore
from haystack.preprocessor.utils import convert_files_to_dicts
from haystack.reader.farm import FARMReader
from haystack.retriever.sparse import TfidfRetriever
from haystack.pipeline import ExtractiveQAPipeline
from haystack.utils import print_answers
```

```python
logging.basicConfig(level=logging.WARNING, force=True)
document_store = InMemoryDocumentStore()
dicts = convert_files_to_dicts(dir_path=WIKIPEDIA_PATH, clean_func=None, split_paragraphs=False)
document_store.write_documents(dicts)
```

```python
reader = FARMReader(model_name_or_path='salti/bert-base-multilingual-cased-finetuned-squad', use_gpu=True)
retriever = TfidfRetriever(document_store=document_store)
pipe = ExtractiveQAPipeline(reader, retriever)
```

```python
def ask(question: str = '', true_answer: str='-'):
    prediction = pipe.run(query=question, top_k_retriever=N_DOCS_CONSIDERED, top_k_reader=5)
    print(f'Question: {question}\nTrue answer: {true_answer}')
    print_answers(prediction, details="medium")
```

```python
df_test = pd.read_csv('./bank_soal/bank_soal.csv')
for question, true_answer in zip(df_test[QUESTION_TYPE], df_test['Jawaban']):
    ask(question, true_answer)
```
