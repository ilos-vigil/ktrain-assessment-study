```python
INDEX_PATH = './index'
WIKIPEDIA_PATH = './dataset/txt'
BATCH_SIZE = 32
N_DOCS_CONSIDERED = 8
RERANK_THRESHOLD = 0.02
QUESTION_TYPE = 'Soal' # "Soal" or "Soal_5W1H"
```

```python
import os
import pandas as pd
from ktrain import text
```

```python
if not os.path.exists(INDEX_PATH):
    print('Index not exist, indexing now...')
    text.SimpleQA.initialize_index(INDEX_PATH)
    text.SimpleQA.index_from_folder(WIKIPEDIA_PATH, INDEX_PATH, procs=6, commit_every=10240, limitmb=1024, multisegment=True, breakup_docs=True)
    print('Indexing done!')

qa = text.SimpleQA(INDEX_PATH, 'salti/bert-base-multilingual-cased-finetuned-squad', 'indobenchmark/indobert-base-p2')
```

```python
def ask(question: str, true_answer: str='-'):
    answers = qa.ask(
        question, batch_size=BATCH_SIZE, n_answers=None, include_np=False,
        n_docs_considered=N_DOCS_CONSIDERED, rerank_threshold=RERANK_THRESHOLD
    )
    print(f'Question: {question}\nTrue answer: {true_answer}')
    qa.display_answers(answers[:5])
```

```python
df_test = pd.read_csv('./dataset/question_sheet.csv')
for question, true_answer in zip(df_test[QUESTION_TYPE], df_test['Jawaban']):
    ask(question, true_answer)
```
