```python
from ktrain import text
from pprint import pprint
```

```python
X_train = [
    'Ubuntu is most popular distro based on Linux',
    'I prefer Debian over Ubuntu',
    'Linux only have 3 percent marketshare',
    'Both Ubuntu and Debian uses systemd'
]
y_train = [
    'Fact',
    'Opinion',
    'Fact',
    'Fact'
]
X_test = [
    'Arch Linux is better if you prefer minimalistic setup',
]
y_test = [
    'Opinion'
]
```

Set maximum unique words to **10**.

```python
train_ds, test_ds, preproc = text.texts_from_array(
    X_train, y_train, X_test, y_test, maxlen=10
)
tokenizer = preproc.get_tokenizer()
print('ID/word dictionary:')
pprint(tokenizer.index_word)
```

As expected, maximum words is 10 and padding is on left of document.

```python
print('Train data:')
print(train_ds[0])
print('Test data:')
print(test_ds[0])
```

Set maximum unique words to **5**.

```python
train_ds, test_ds, preproc = text.texts_from_array(
    X_train, y_train, X_test, y_test, maxlen=5
)
```

As expected, maximum words is 5. We also can see that beginning of the document is removed.

```python
print('Train data:')
print(train_ds[0])
print('Test data:')
print(test_ds[0])
```
