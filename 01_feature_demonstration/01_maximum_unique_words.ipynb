```python
from ktrain import text
from pprint import pprint
```

Here's example document to show how ktrain select unique words. Take note word "Ubuntu", "Linux" and "Debian" occured more than once on training document.

```python
X_train = [
    'Ubuntu is most popular distro based on Linux',
    'I prefer Debian over Ubuntu',
    'Linux only have 3 percent marketshare',
    'Both Ubuntu and Debian uses systemd'
]
y_train = [
    'Fact',
    'Opinion',
    'Fact',
    'Fact'
]
X_test = [
    'Arch Linux is better if you prefer minimalistic setup',
]
y_test = [
    'Opinion'
]
```

Set maximum unique words to **5**.

```python
train_ds, test_ds, preproc = text.texts_from_array(
    X_train, y_train, X_test, y_test,
    max_features=5, maxlen=10
)
```

Obtain `tokenizer` from `preproc` object, then print it's ID/word dictionary.

```python
tokenizer = preproc.get_tokenizer()
print('ID/word dictionary:')
pprint(tokenizer.index_word)
```

Looking at preprocessed document, there are 5 unique ID. We can see that ID `0` is used as padding value. Additionally words with highest frequency followed by words which appear first is priotized.

```python
print('Train data:')
print(train_ds[0])
print('Test data:')
print(test_ds[0])
```
