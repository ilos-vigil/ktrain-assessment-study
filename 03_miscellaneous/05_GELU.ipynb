```python
import torch
import numpy as np

input_ = torch.tensor([
    [0.1, -0.5, 0.7, -0.2],
    [-0.2, 0.2, 0.6, -0.3]
])
```

```python
out_ = torch.nn.functional.gelu(input_)
out_
```

```python
def gelu(x):
    y = 0.5 * x * (
        1 + np.tanh(
            np.sqrt(2 / np.pi) * (
                x + 0.044715 * np.power(x,3)
            )
        )
    )
    return y
```

```python
out_ = input_.detach().numpy()
gelu_ = np.vectorize(gelu)
out_ = gelu_(out_)
out_
```
