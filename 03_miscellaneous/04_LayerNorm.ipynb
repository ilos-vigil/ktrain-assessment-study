```python
import torch
from math import sqrt
from collections import OrderedDict
```

```python
input_data = torch.tensor([0.1, 0.3, -0.4, 0.6])
weight = torch.tensor([0.5, -0.25, 0.1, 0.2])
bias = torch.tensor([-0.03, 0.01, 0.04, -0.08])
epsilon = 0.001

param = OrderedDict({
    'weight': weight,
    'bias': bias,
})
```

```python
ln = torch.nn.LayerNorm((4), eps=epsilon)
ln.load_state_dict(param, strict=False)
ln(input_data)
```

```python
k = len(input_data)

mean_i = sum(input_data[j] for j in range(k)) / k
print(f'{mean_i=}')
var_i = sum((input_data[j] - mean_i) ** 2 for j in range(k)) / k
print(f'{var_i=}')
x_i_normalized = (input_data - mean_i) / sqrt(var_i + epsilon)
print(f'{x_i_normalized=}')
output_i = x_i_normalized * weight + bias
print(f'{output_i=}')
```
